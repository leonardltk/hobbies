```python
import os
import cv2
import pdb
import base64
import traceback
from rich import print as rich_print
from dotenv import load_dotenv, find_dotenv

from openai import OpenAI
```

```python
load_dotenv(find_dotenv())
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")

client = OpenAI()
```

```python
# read file
file = client.files.create(
    file=open("data/Film Classification Guidelines 29_Apr_2019.pdf", "rb"),
    purpose="user_data"
)
```

```python
# Generate meta prompt
response = client.responses.create(
    model="gpt-5",
    input=[
        {
            "role": "user",
            "content": [
                {
                    "type": "input_file",
                    "file_id": file.id,
                },
                {
                    "type": "input_text",
                    "text": (
                        "Generate a system prompt for me to make a classification of a movie that will subsequently "
                        "be passed into the model to classify a rating. "
                        "Suggest structured output format too"
                    ),
                },
            ]
        }
    ]
)
print(response.output_text)
```
